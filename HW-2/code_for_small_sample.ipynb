# 匯入資料及前處理

```python
import pandas as pd
import numpy as np

# pd.set_option('precision', 0)
data = pd.read_csv('input.txt', sep="\t",header = None, error_bad_lines=False)
```

```python
data = data.drop(columns = 0,axis = 1)
```

```python
#將資料從dataframe轉為array
test = data.head(10)
test = test.values
test
```

```python
#將資料從array轉為二維list
import numpy as np
import math
news = []
for i in test:
    tmp = [x for x in i if math.isnan(x) == False]
    news.append(tmp)
new = []
for i in news:
    tmp = [int(x) for x in i if math.isnan(x) == False]
    new.append(tmp)
new
```

```python
new_data=[] #取出資料欄位並創建新的list存放
for i in new:
    tmp=[]
    for j in range(2,len(i)+1,2):
        tmp.append(i[j-1])
    new_data.append(tmp)
new_data
```

```python
index = [] #建立list存放transaction time
for i in new:
    tmp=[]
    for j in range(0,len(i),2):
        tmp.append(i[j])
    index.append(tmp)
index
```

# 計算次數用於資料前處理(把同時間之資料合併)

```python
inits = [] #紀錄每一種資料
datas = []

for i in index:
    tmp = []
    for j in i:       # 1st loop
        if j not in tmp:   # 2nd loop
            tmp.append(j)
    datas.append(tmp)
for i in datas:
    for j in i:
        if(j not in inits):
            inits.append(j)
print(inits)
```

```python
from collections import Counter #匯入一種dict的子類別 存放資料
p = Counter()
for i in inits:
    for d in index:
        for x in d:
            if(i == x):
                p[i]+=1
print("計算數量：") #每種資料個紀錄一次(出現次數)
for i in p:
    print(str([i])+": "+str(p[i])) #出現次數
print()
```

```python
new_new_data = []
for i in new_data:
    for j in i:
        new_new_data.append(j)
new_new_data
```

# 完成刪除不符合min_sup之資料

```python
teemp = []
for i in index_ok:
    tmp = []
    for j in i:
        if len(j) != 0:
            tmp.append(j)
    if len(tmp) > 1:
        teemp.append(tmp)
teemp
```

# 處理及列舉所有符合min_sup之交易序列

```python
combinations = []
import itertools
for my_list in teemp:
    for i in range(len(my_list)):
        for combination in itertools.product(*my_list[:i+1]):
            if i == len(my_list) - 1:
#                 print([[x] for x in combination])
                combinations.append([[x] for x in combination])
            else:
#                 print(my_list[:i] + [list(combination)] + my_list[i+1:])
                combinations.append(my_list[:i] + [list(combination)] + my_list[i+1:])
combinations
```

```python
combinations_unique = [] #完成之資料序列
for i in combinations:         # 1st loop
    if i not in combinations_unique and len(i) > 1:   # 2nd loop
        combinations_unique.append(i)
combinations_unique
```
